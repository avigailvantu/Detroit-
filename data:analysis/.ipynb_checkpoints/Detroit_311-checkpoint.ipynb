```python
# remove first charc
det_prob['Location'] = det_prob['Location'].map(lambda x: str(x)[10:])
```

```python
# remove last charc
det_prob['Location']  = det_prob['Location'] .map(lambda x: str(x)[:-1])
```

```python
#det_prob['Location'] = det_prob[['Location'].map(lambda x: x.lstrip('location\n').rstrip('aAbBcC'))
```

```python
lat = []
lon = []

# For each row in a varible,
for row in det_prob['Location']:
    # Try to,
    try:
        # Split the row by comma and append
        # everything before the comma to lat
        lat.append(row.split(',')[0])
        # Split the row by comma and append
        # everything after the comma to lon
        lon.append(row.split(',')[1])
    # But if you get an error
    except:
        # append a missing value to lat
        lat.append(np.NaN)
        # append a missing value to lon
        lon.append(np.NaN)

# Create two new columns from lat and lon
det_prob['latitude'] = lat
det_prob['longitude'] = lon
```
